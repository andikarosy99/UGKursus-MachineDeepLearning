```python
from boilerpipe.extract import Extractor
from gensim import corpora
from pprint import pprint
import string

from gensim.models import LdaModel
from nltk.tokenize import word_tokenize,sent_tokenize,WhitespaceTokenizer

import pyLDAvis.gensim

import warnings
warnings.filterwarnings('ignore')
#import pymysql.cursors
import io
```

```python
f = io.open("data/testing-ringkasan/full", mode="r", encoding="utf-8")
#f = open("data/testing-ringkasan/full")
renstra = f.read()
content=renstra
content
```

```python
d = open("data/StopWordID2")
stopword_id = d.read()
indonesia_stopwords = word_tokenize(stopword_id)
#indonesia_stopwords

punctuation = string.punctuation
additional_punctuation = [".", "-", "``","– ",":","meningkatnya","1.","2.","3.","4.","5.","sasaran:","program:","v","6.","2013-2017"]
array_punctuation = word_tokenize(punctuation)
stopword = indonesia_stopwords + array_punctuation + additional_punctuation
stopword
```

```python
texts = [[word for word in content.lower().split() if word not in stopword]]
corpus_frequensi = [word for word in content.lower().split() if word not in stopword]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
texts
import nltk
counts  = nltk.FreqDist(corpus_frequensi)
counts
#mydict = {'carl':40,
#          'alan':2,
#          'bob':1,
#          'danny':3}
#print (len(counts))
#for key, value in sorted(counts.iteritems(),reverse=True, key=lambda (k,v): (v,k)):
#    print ("%s: %s" % (key, value))
    
```

```python
# menghitung nilai lda dari model LDA
ldamodel = LdaModel(corpus=corpus, num_topics=10, id2word=dictionary)
data = ldamodel.show_topics()
data[0]
```

```python
pyLDAvis.enable_notebook()
pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```
