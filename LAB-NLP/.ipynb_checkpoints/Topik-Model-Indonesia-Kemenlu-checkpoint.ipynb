## 1. Crawling atau Akusisi Data

```python
#import pyLDAvis.gensim
#from gensim import corpora, models, similarities 
#from gensim.models import hdpmodel, LdaModel 
from gensim import corpora,models
from pprint import pprint
import matplotlib.pyplot as plt
import gensim
import numpy as np
import spacy

from gensim.models import CoherenceModel, LdaModel, LsiModel, HdpModel
from gensim.models.wrappers import LdaMallet
from gensim.corpora import Dictionary
import pyLDAvis.gensim
#from itertools import izip
documents = [] 
isiText = ''
#Tema1.txt adalah file teks dari tema1
with open("data/data-kemenlu.txt", "r") as f:   
	#for line in f:   
	while 1:
		isi = f.readline()
		if isi == '':         
			break 
		isiText += isi   	    		
	
#documents.append(line.splitlines()) 
documents.append(isiText) 
print (documents)

```

## 2. Stop Word List (Stop-Removal)

```python
# remove common words and tokenize 
stoplist = set('dan yang di ke the of and to on oleh : for dengan dari pada untuk dalam tersebut telah in  akan tanggal negeri *'.split()) 
```

```python
#stoplist = set('for a of the and to in'.split())
texts = [[word for word in document.lower().split() 
		if word not in stoplist]          		
for document in documents] 
print (texts)
```

## 3. Pembuatan Corpus

```python
dictionary = corpora.Dictionary(texts) 
corpus = [dictionary.doc2bow(text) for text in texts] 

tfidf = models.TfidfModel(corpus) 
corpus_tfidf = tfidf[corpus] 
```

```python
#menghitung nilai lda dari model LDA
ldamodel = LdaModel(corpus=corpus, num_topics=20, id2word=dictionary)
```
