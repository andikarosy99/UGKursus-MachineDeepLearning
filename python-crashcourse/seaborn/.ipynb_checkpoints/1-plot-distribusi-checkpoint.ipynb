# Distribution Plots

Notebook ini berisi tentang cara menggunakan Library Seaborn untuk melakukan visualisasi data yang berupa sebaran atau distribusi.

___
## Imports

```python
import seaborn as sns
%matplotlib inline
```

## Data

Salah satu kelebihan library seaborn adalah adanya built-in dataset. Jadi kita bisa langsung belajar cara melakukan visualisasi tanpa mengkhawatirkan data.

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips.head()
```

## distplot

Fungsi distplot menunjukkan distribusi himpunan univariate dari observasi.

```python
sns.distplot(tips['total_bill'])
# Safe to ignore warnings
```

Jika tidak ditetapkan, Kernel Density Estimation (KDE) akan secara otomatis muncul dalam visualisasi distribusi kita. Kita bisa tetapkan **KDE=False** untuk menghilakannya. Kita juga bisa mengatur jumlah bin dari distribusi, silahkan mencoba untuk mengubah-ubah parameter tersebut.

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jointplot

Fungsi jointplot pada dasarnya menggabungkan dua buah displot untuk menunjukkan visualisasi data bivariate. Ada beragam pilihan **kind** yang bisa dipilih:
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

## pairplot

Fungsi pairplot akan melakukan plotting relasi keseluruhan dataframe, memberikan keleluasaan kita untuk menganilisis data kita lebih dalam.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```
